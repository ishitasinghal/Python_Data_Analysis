### Import required pandas library.

```python
import pandas as pd
```

### Import the data set from [this]('https://raw.githubusercontent.com/datacamp/datacamp_facebook_live_ny_resolution/master/data/multiTimeline.csv') location 

```python
data = pd.read_csv('multiTimeline.csv', skiprows=1)
data.head()
```

### Print first 5 rows for the data.

```python
data.head()
```

### Change the columns name to Diet, Gym, Finance

```python
data.columns = ['Month', 'Diet', 'Gym', 'Finance']
data.head()
```

### Print the data type of each column

```python
data.info()
```

### Change the Month columns data type to have date time features

```python
data['Month'] = pd.to_datetime(data['Month'])
```

### Make the Month column as the index of the dataframe.

```python
data.info()
```

```python
data = data.set_index('Month')
data.head()
```

## Plot the data using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 8))
plt.plot(data.Diet)
plt.plot(data.Gym)
plt.plot(data.Finance)
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.legend(loc='upper right')
plt.title('Trend each year')
plt.show()
```

## Smoothe the data, to see the trend in data. 

```python
plt.figure(figsize=(10, 8))
plt.plot(data.Diet.rolling(12).mean())
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.legend(loc='upper right')
plt.title('Dieting Trend each year')
plt.savefig('Diet Trend.png')
plt.show()
```

## Repeat the smoothing procedure for Gym, Finance

```python
plt.figure(figsize=(10, 8))
plt.plot(data.Gym.rolling(12).mean())
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.legend(loc='upper right')
plt.title('Gym Trend each year')
plt.savefig('Gym Trend.png')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(data.Finance.rolling(12).mean())
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.legend(loc='upper right')
plt.title('Finance Trend each year')
plt.savefig('Finance Trend.png')
plt.show()
```

## Compare the trend of Gym and Finance

```python
plt.figure(figsize=(10, 8))
plt.plot(data.Gym.rolling(12).mean())
plt.plot(data.Finance.rolling(12).mean())
plt.xlabel('Year')
plt.ylabel('Number of people')
plt.legend(loc='upper right')
plt.title('Dieting Trend each year')
plt.savefig('Diet Trend.png')
plt.show()
```
